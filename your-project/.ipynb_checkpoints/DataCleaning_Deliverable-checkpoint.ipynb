```python
# since I want to correlate things with revenue, it doesn't make sense to have records with no revenue info, so I just dropped
# those rows

# drop rows with no info on revenue
IMDB_drop_nan = data_IMDB.dropna(subset=['Revenue (Millions)'])
IMDB_drop_nan.head()
```

```python
# check for missing values
IMDB_drop_nan.isnull().sum()
```

```python
# now I only have NaN values in Metascore, let's check

NaN_metascore = IMDB_drop_nan[IMDB_drop_nan.Metascore.isnull()]
NaN_metascore.head()

# I'll keep these rows to perform the analysis and only not consider them when relating things with metascore
```

```python
# check if the NaN values match low votting
NaN_metascore.sort_values(by=['Votes'])
```

```python
IMDB_drop_nan.describe()
```

```python
# votes range from 178 to 1 791 916 and the average is 134 654. As a criterion, I decided to consider only movies with
# 30 000 or more votes. This is IMDB's minimum votes value for a movie to be in top250
min_votes = 30000
IMDB_min_votes = IMDB_drop_nan[IMDB_drop_nan.Votes >= min_votes]
IMDB_min_votes.describe()
```

```python
# getting an overview of the new df
IMDB_min_votes.shape
```

```python
# reset index
IMDB_index = IMDB_min_votes.reset_index(drop=True)
```

```python
# confirm the reindexing
IMDB_index.index
```

```python
# In the beginning, I wanted to relate revenue and director genre
# I started by creating a list of the director's unique first names

# insert director genre
dir_lst_uniques = sorted(list(IMDB_index.Director.unique()))
len(dir_lst_uniques)

dir_first_name = sorted(set([dir_lst_uniques[i].split()[0] for i in range(len(dir_lst_uniques))]))
        
#print(dir_first_name)
len(dir_first_name)
```

```python
'''
With the unique names list, I used  the Gender Api (https://gender-api.com/en/)
to get the genre name.
'''
```

```python
# getting the names gender data
dir_names_genre = pd.read_csv(r'C:\Users\Ana André\Ironhack\projectos\projecto4\first_name_gender.csv',sep=';')
dir_names_genre.head()
```

```python
dir_names_genre.describe()
```

```python
''' The data from the Gender Api has info about the accuracy and samples used to
decide the genre of a name.
To check the accuracy of the genre api results, I decide to double check the name
records where accuracy and number of samples were less than 50% and 1000, respectively
'''
# sort genre names df by the accuracy percentage
dir_names_genre.sort_values(by=['ga_accuracy'])[:10]
```

```python
# list the names to check by hand
dir_names_to_check = dir_names_genre[(dir_names_genre['ga_samples'] < 1000) | (dir_names_genre['ga_accuracy'] < 50)]
dir_names_to_check.head()
len(dir_names_to_check)
```

```python
# create a dictionary of the names and genre
names_gender_dict = dict(zip(dir_names_genre['Aamir'].to_list(),dir_names_genre['ga_gender'].to_list()))
#print(names_gender_dict)
```

```python
# strip the names/genre dictionary
names_gender_dict_strip = [name.strip() for name in names_gender_dict.keys()]
#print(names_gender_dict_strip)
```

```python
# add new columns to the df: director's first and last name
split_dir_name = IMDB_index['Director'].str.split(" ", n = 1, expand = True)
IMDB_index['Director_Fname'] = split_dir_name[0]
IMDB_index['Director_Lname'] = split_dir_name[1]
IMDB_index.tail()
```

```python
# create a list with the director's genre based on the first name
dir_genre = []

#print(len(IMDB_index['Director_Fname']))

for name in IMDB_index['Director_Fname']:
    #print(name)
    if name in names_gender_dict_strip:
        #print(name)
        x = names_gender_dict.get(" "+ name)
        #print(x)
        dir_genre.append(x)
        
print(len(dir_genre))

# By printing the length, I noticed that I've a missing values, I was expecting
# 756 items.
```

```python
'''
The missing value is the name 'Aamir' which is wrongly set as a column
label in the director_name_genre df.
I decided to add this director's genre by hand because it was only one
value.
'''

# check for director's "Aamir Khan" index in the dataframe
IMDB_index[IMDB_index.Director == 'Aamir Khan'].index[0]
```

```python
# add the director's "Aamir Khan" genre to the director's genre list
dir_genre.insert(IMDB_index[IMDB_index.Director == 'Aamir Khan'].index[0],'male')

print(len(dir_genre))

# now I have all the values.
```

```python
# create the director gender column
IMDB_index['Director_genre'] = dir_genre
IMDB_index.head()
```

```python
# Having the dataset ready for the analysis, I saved it as a csv file so I
# could work on another file.
IMDB_index.to_csv('Movies_DB_clean.csv')
```

```python
# creating the correlation matrix
corr_matrix = IMDB_index[['Rank','Year','Rating','Votes','Revenue (Millions)']].corr()
corr_matrix
```

```python
# ploting the correlation matrix
sns.set(style="white")

# generate a mask for the upper triangle
mask = np.zeros_like(corr_matrix, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True

# set up the matplotlib figure
f, ax = plt.subplots(figsize=(11,9))

# generate a custom diverging colormap
cmap = sns.diverging_palette(220,10,as_cmap = True)

# draw the heatmap with the mask and correct aspect ration
sns.heatmap(corr_matrix, mask=mask, cmap = cmap, vmax=0.61 , center = 0,
           square = True, linewidths=.5, cbar_kws={'shrink':.5})

plt.show()
```
