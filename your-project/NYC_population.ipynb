```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv('NYC_Population_by_Borough.csv')
ds.head(10)
```

```python
ds['Population'].sum()
```

```python
ds['Population'].astype('float')
```

```python
labels = ['\nBrooklyn', 'Manhattan', 'Bronx', 'Queens', 'Staten Island']
```

```python
sns.set(font_scale = 1.6)
plt.style.use('dark_background')
plt.figure(figsize=(9,5))
colors = ['green', 'blue', 'orange', 'grey', 'red']
explode = (0.2, 0.0, 0.0, 0.0, 0.0)
plt.pie(ds['Population'], explode=explode, labels=labels, colors = colors,
        autopct='%1.0f%%', shadow=True, startangle=65) 
plt.axis('equal')
plt.title('Population per borough in NYC (2019)')

plt.savefig('pop_borough.png', bbox_inches = 'tight', transparent = True)
```

```python
ds_pop = pd.read_csv('New_York_City_Population_by_Borough__1950_-_2040.csv')
ds_pop.head(10)
```

```python
ds_pop = ds_pop.drop(ds_pop[ds_pop['Borough'] == 'NYC Total'].index)
```

```python
sns.set(style="whitegrid", font_scale = 1.5)
plt.figure(figsize=(8, 5))

ax = sns.barplot(x='2020', y='Borough', data=ds_pop, saturation=1.5)
ax1 = sns.barplot(x='2030', y='Borough', data=ds_pop, saturation=1.5)
ax2 = sns.barplot(x='2040', y='Borough', data=ds_pop, saturation=1.5)

ax = plt.xticks(rotation=45)
plt.title("Population per Borough projection")
plt.xlabel('Population')
plt.ylabel('Borough')
plt.show()
```

```python
# reshaping dataframe to have desired bar plot
df = pd.melt(ds_pop, id_vars='Borough', value_vars=['2020','2030','2040'])
df
```

```python
sns.set(font_scale = 1.5)
plt.style.use('dark_background')
plt.figure(figsize=(8, 5))

ax = sns.barplot(x='value', y='Borough', hue= 'variable', data=df,
                palette='muted')

ax = plt.xticks(rotation=45)
plt.title("Population per Borough projection")
plt.legend(loc=0, handlelength=0.5)
plt.xlabel('Population')
plt.ylabel('Borough')

plt.savefig('pop_project.png', bbox_inches = 'tight', transparent = True)
```
